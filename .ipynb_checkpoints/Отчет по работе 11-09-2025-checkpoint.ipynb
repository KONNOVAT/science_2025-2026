# Отчет по исследованию системы конкуренции видов

## Введение

## Теоретическая основа

В данном отчете рассматривается математическая модель конкуренции двух видов за общие ресурсы. Модель описывается системой дифференциальных уравнений для конкуренции за ресурсы:

du₁/dt = u₁ * (1 - u₁ - ε₁ * u₂)
du₂/dt = u₂ * (1 - u₂ - ε₂ * u₁)

**где:**

**u₁, u₂** - численности видов 1 и 2 (нормированные на емкость среды)

**ε₁, ε₂** - коэффициенты конкуренции, показывающие влияние одного вида на другой

## ИССЛЕДОВАНИЕ СИСТЕМЫ КОНКУРЕНЦИИ ВИДОВ

```julia
using Plots
using LinearAlgebra

# Создаем папку для графиков
# Это обеспечивает организованное хранение всех результатов
graphics_dir = "graphics"
if !isdir(graphics_dir)
    mkdir(graphics_dir)
    println("Создана папка '$graphics_dir'")
end

"""
    competition_system!(X, p)

Система уравнений конкуренции двух видов.
Основана на модели Лотки-Вольтерры для конкуренции.

# Аргументы
- `X`: вектор [u₁, u₂] текущих численностей
- `p`: параметры [ε₁, ε₂] - коэффициенты конкуренции

# Уравнения
du₁/dt = u₁ * (1 - u₁ - ε₁ * u₂)
du₂/dt = u₂ * (1 - u₂ - ε₂ * u₁)
"""
function competition_system!(X, p)
    ε1, ε2 = p
    u1, u2 = X
    du1 = u1 * (1 - u1 - ε1 * u2)  # Рост вида 1 с учетом внутривидовой и межвидовой конкуренции
    du2 = u2 * (1 - u2 - ε2 * u1)  # Рост вида 2 с учетом внутривидовой и межвидовой конкуренции
    return [du1, du2]
end
```

**solve_competition(u0, ε1, ε2, T=20.0, dt=0.1)**

## Численное решение системы конкуренции методом Эйлера.

## Аргументы
- `u0`: начальные условия [u₁₀, u₂₀]
- `ε1, ε2`: параметры конкуренции
- `T`: время моделирования
- `dt`: шаг по времени

## Возвращает
- `times`: массив временных точек
- `U`: матрица решений [u₁(t), u₂(t)] для каждого времени

```julia
function solve_competition(u0, ε1, ε2, T=20.0, dt=0.1)
    times = 0:dt:T
    n_steps = length(times)
    U = zeros(n_steps, 2)
    U[1, :] = u0  # Устанавливаем начальные условия

    # Численное интегрирование методом Эйлера
    for i in 2:n_steps
        du = competition_system!(U[i-1, :], (ε1, ε2))
        U[i, :] = U[i-1, :] + dt * du  # Простейшая схема Эйлера
    end

    return times, U
end
```

- plot_phase_portrait(ε1, ε2, scenario_name)

Построение фазового портрета системы.
Фазовый портрет показывает траектории системы в пространстве состояний (u₁, u₂).

# Особенности
- Несколько начальных условий для демонстрации разных траекторий
- Цветовое кодирование траекторий
- Маркировка начальных и конечных точек

```julia
function plot_phase_portrait(ε1, ε2, scenario_name)
    # Разнообразные начальные условия для демонстрации разных сценариев
    initial_conditions = [
        [0.1, 0.1],  # Оба вида с низкой численностью
        [0.8, 0.1],  # Вид 1 доминирует изначально
        [0.1, 0.8],  # Вид 2 доминирует изначально
        [0.4, 0.4],  # Равные начальные условия
        [0.7, 0.7]   # Оба вида с высокой численностью
    ]

    # Создаем основу для графика
    p = plot(xlabel="Вид 1 (u₁)", ylabel="Вид 2 (u₂)",
             title="Фазовый портрет: $scenario_name",
             legend=:topright, size=(600, 600), grid=true,
             xlim=(0, 1), ylim=(0, 1))

    colors = [:blue, :red, :green, :purple, :orange]

    # Рисуем траектории для каждого начального условия
    for (idx, u0) in enumerate(initial_conditions)
        times, U = solve_competition(u0, ε1, ε2)

        # Основная траектория
        plot!(p, U[:, 1], U[:, 2], linewidth=2.5,
              color=colors[idx], label="начало: ($(u0[1]), $(u0[2]))")

        # Начальная точка (зеленая)
        scatter!(p, [u0[1]], [u0[2]], color=colors[idx], markersize=5)

        # Конечная точка (красный квадрат)
        scatter!(p, [U[end, 1]], [U[end, 2]], color=colors[idx],
                 markersize=6, marker=:square)
    end

    return p
end
```

**create_simple_animation(ε1, ε2, scenario_name)**

Создание анимации для одной траектории.
Анимация показывает одновременную эволюцию:
- Временных рядов численностей
- Фазовой траектории

# Особенности анимации
- Синхронизированное отображение времени на обоих графиках
- Динамическое обновление траекторий
- Визуализация текущего состояния системы

```julia
function create_simple_animation(ε1, ε2, scenario_name)
    u0 = [0.2, 0.8]  # Начальные условия: вид 2 доминирует
    T = 10.0  # Укороченное время для более быстрой анимации
    dt = 0.1
    times, U = solve_competition(u0, ε1, ε2, T, dt)

    # Используем каждый второй кадр для оптимизации
    frame_indices = 1:2:length(times)

    anim = @animate for i in frame_indices
        current_time = times[i]

        # Создаем компоновку из двух графиков
        p = plot(layout=(1,2), size=(1000, 400))

        # ЛЕВЫЙ ГРАФИК: Динамика численностей во времени
        plot!(p[1], times[1:i], U[1:i, 1], linewidth=3, color=:blue,
              label="Вид 1 (u₁)", xlabel="Время", ylabel="Численность",
              title="Динамика: $scenario_name",
              xlim=(0, T), ylim=(0, 1.2), grid=true, legend=:topright)

        plot!(p[1], times[1:i], U[1:i, 2], linewidth=3, color=:red,
              label="Вид 2 (u₂)")

        # Маркер текущего времени
        vline!(p[1], [current_time], color=:black, linestyle=:dash, label="")
        annotate!(p[1], current_time, 1.1, text("t = $(round(current_time, digits=1))", 10))

        # ПРАВЫЙ ГРАФИК: Фазовый портрет
        plot!(p[2], U[1:i, 1], U[1:i, 2], linewidth=3, color=:purple,
              label="Траектория", xlabel="Вид 1 (u₁)", ylabel="Вид 2 (u₂)",
              title="Фазовый портрет",
              xlim=(0, 1), ylim=(0, 1), grid=true, legend=:topright)

        # Начальная и текущая точки
        scatter!(p[2], [u0[1]], [u0[2]], color=:green, markersize=6,
                label="Начало")
        scatter!(p[2], [U[i, 1]], [U[i, 2]], color=:red, markersize=8,
                marker=:square, label="Текущее")
    end

    return anim
end
```

**create_multi_animation(ε1, ε2, scenario_name)**

Создание анимации с несколькими траекториями.
Демонстрирует зависимость конечного состояния от начальных условий.

# Особенности
- Три различные начальные точки
- Параллельное отображение временной динамики и фазовых траекторий
- Сравнительный анализ разных сценариев

```julia
function create_multi_animation(ε1, ε2, scenario_name)
    initial_conditions = [[0.2, 0.2], [0.8, 0.2], [0.2, 0.8]]
    T = 8.0  # Оптимизированное время для анимации
    dt = 0.1
    colors = [:blue, :red, :green]

    # Решаем систему для всех начальных условий
    solutions = []
    for u0 in initial_conditions
        times, U = solve_competition(u0, ε1, ε2, T, dt)
        push!(solutions, (times, U))
    end

    # Используем каждый третий кадр для баланса плавности и производительности
    frame_indices = 1:3:length(solutions[1][1])

    anim = @animate for i in frame_indices
        current_time = solutions[1][1][i]

        p = plot(layout=(1,2), size=(1000, 400))

        # ЛЕВЫЙ ГРАФИК: Временная динамика всех траекторий
        for (idx, (times, U)) in enumerate(solutions)
            plot!(p[1], times[1:i], U[1:i, 1], linewidth=2,
                  color=colors[idx], label="u₁ траек. $idx",
                  xlabel="Время", ylabel="Численность",
                  title="Динамика: $scenario_name",
                  xlim=(0, T), ylim=(0, 1.2), grid=true, legend=:right)
            plot!(p[1], times[1:i], U[1:i, 2], linewidth=2,
                  color=colors[idx], label="u₂ траек. $idx", linestyle=:dash)
        end

        # ПРАВЫЙ ГРАФИК: Фазовые траектории
        for (idx, (times, U)) in enumerate(solutions)
            plot!(p[2], U[1:i, 1], U[1:i, 2], linewidth=2,
                  color=colors[idx], label="траектория $idx",
                  xlabel="Вид 1 (u₁)", ylabel="Вид 2 (u₂)",
                  title="Фазовый портрет",
                  xlim=(0, 1), ylim=(0, 1), grid=true, legend=:topright)
            scatter!(p[2], [initial_conditions[idx][1]], [initial_conditions[idx][2]],
                    color=colors[idx], markersize=4)
        end
    end

    return anim
end
```

**plot_population_dynamics(ε1, ε2, scenario_name)**

Построение графиков временной динамики численностей.
Показывает эволюцию системы для различных начальных условий.

# Визуальные особенности
- Сплошные линии для вида 1, пунктирные для вида 2
- Цветовое кодирование по начальным условиям
- Легенда с указанием начальных значений


```julia
function plot_population_dynamics(ε1, ε2, scenario_name)
    initial_conditions = [
        [0.2, 0.2], [0.8, 0.2], [0.2, 0.8], [0.6, 0.6]
    ]

    p = plot(xlabel="Время", ylabel="Численность",
             title="Динамика популяций: $scenario_name",
             legend=:right, size=(600, 400), grid=true)

    colors = palette(:default)

    for (i, u0) in enumerate(initial_conditions)
        times, U = solve_competition(u0, ε1, ε2)

        # Вид 1 - сплошная линия
        plot!(p, times, U[:, 1], linewidth=2,
              color=colors[i], label="u₁ (начало: $(u0[1]),$(u0[2]))", linestyle=:solid)
        # Вид 2 - пунктирная линия
        plot!(p, times, U[:, 2], linewidth=2,
              color=colors[i], label="u₂ (начало: $(u0[1]),$(u0[2]))", linestyle=:dash)
    end

    return p
end
```

**save_plot(plot, filename)**

Утилита для сохранения графиков в папку graphics.

```julia
function save_plot(plot, filename)
    full_path = joinpath(graphics_dir, filename)
    savefig(plot, full_path)
    println("Сохранен: $full_path")
end
```
